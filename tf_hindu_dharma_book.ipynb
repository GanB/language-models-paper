```python
path="/content/drive/MyDrive/Colab Notebooks/data/hindu-dharma/hindu-dharma.txt"
```

```python
hindu_dharma_txt = open(path, encoding="utf8")
#print(hindu_dharma_txt.read())
hindu_dharma_data = hindu_dharma_txt.read()
hindu_dharma_words = hindu_dharma_data.replace("\n", " ").split(" ")
hindu_dharma_sentences = hindu_dharma_data.replace("\n", " ").split(".")
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

print("TensorFlow version: ", tf.__version__)
```

```python
tokenizer = Tokenizer(num_words=len(hindu_dharma_words), oov_token="<OOV>")
tokenizer.fit_on_texts(hindu_dharma_sentences)
word_index = tokenizer.word_index
print(word_index)
```

```python
word_index
```

```python
sequences = tokenizer.texts_to_sequences(hindu_dharma_sentences)

padded = pad_sequences(sequences, maxlen=5)
print("\nWord Index = " , word_index)
print("\nSequences = " , sequences)
print("\nPadded Sequences:")
print(padded)

test_data = ["what are sastra?", "what are mandapa?"]

test_seq = tokenizer.texts_to_sequences(test_data)
print("\nTest Sequence = ", test_seq)

padded = pad_sequences(test_seq, maxlen=10)
print("\nPadded Test Sequence: ")
print(padded)
```
